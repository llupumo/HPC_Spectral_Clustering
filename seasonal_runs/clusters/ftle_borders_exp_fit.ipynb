```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Define the exponential decay function
def exp_decay(x, a, b, c):
    return a * np.exp(-b * x) + c

# Define the derivative of the fitted function
def exp_decay_derivative(x, a, b):
    return -a * b * np.exp(-b * x)

# Function to compute the flat point
def find_flat_point(year, season, threshold=0.0001, plot=True):
    """
    Finds the flat point of the mean FTLE for a given year and season.
    Parameters:
        year (int): The year of the data.
        season (str): The season (e.g., "AMJ", "JJA").
        threshold (float): The flatness threshold for the derivative.
        plot (bool): Whether to plot the data and the fit.
    Returns:
        float: The flat point where the function is "flat enough."
    """
    # Load the FTLE data
    file_path = f"/cluster/projects/nn8008k/lluisa/NextSIM/seas/Fmap/{year}_{season}_ic0.5_dt0.0025_DT0.1000/clusters_K10000000_border0.00/2_100_FTLE.npy"
    FTLE = np.load(file_path, allow_pickle=True)
    # Compute mean FTLE and indices
    mean_FTLE = []
    idx = []
    for i in range(len(FTLE)):
        mean_FTLE.append(np.mean(np.abs(FTLE[i])))
        idx.append(2 + i)
    # Fit the exponential decay function to the data
    params, _ = curve_fit(exp_decay, idx, mean_FTLE, p0=[10, 0.5, 1])
    a, b, c = params
    print(f"Fitted parameters: a = {a:.3f}, b = {b:.3f}, c = {c:.3f}")
    # Find the flat point
    flat_point = None
    for x_val in idx:
        if abs(exp_decay_derivative(x_val, a, b)) < threshold:
            flat_point = x_val
            break
    print(f"The function is flat enough at x = {flat_point:.2f}")
    # Plot the data and the fit if requested
    if plot:
        x_fit = np.linspace(min(idx), max(idx), 500)
        y_fit = exp_decay(x_fit, a, b, c)
        plt.scatter(idx, mean_FTLE, label='Data', color='blue')
        plt.plot(x_fit, y_fit, label=f'Fit: $y = {a:.3f} e^{{-{b:.3f} x}} + {c:.3f}$', color='red')
        plt.axvline(flat_point, color='green', linestyle='--', label=f'Flat Point: x = {flat_point:.2f}')
        plt.title(f'Exponential Decay Fit and Flatness ({year} {season})',fontsize=14)
        plt.legend(fontsize=14)
        plt.xticks(fontsize=12)
        plt.yticks(fontsize=12)
        plt.xlabel("Number of clusters",fontsize=16)
        plt.ylabel(r' $\overline{\text{FTLE}}$ in boundaries',fontsize=16)
   


        plt.legend()
        plt.grid(True)
        plt.show()
    return flat_point
```
