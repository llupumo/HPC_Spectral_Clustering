```python
from itertools import combinations

def cut_weight_calculator(labels):
    cluster_cut_weight = np.empty(len(set(labels)))
    ic_cut_weight = np.empty(len(labels))

    for k in set(labels):
        positions_k = [index for index, value in enumerate(labels) if value == k]
        W_tri = np.triu(W,k=1)
        pairs = combinations(positions_k, 2)  # itertools.combinations ensures i < j
        # Set matrix[i, j] to 0 for each pair
        for i, j in pairs:
            W_tri[i, j] = 0
        cluster_cut_weight[k]=np.sum(W_tri)

    for i in range(len(labels)):
        ic_cut_weight[i] = cluster_cut_weight[labels[i]]

    return ic_cut_weight
```

```python
from itertools import combinations, product

def cut_weight_calculator(labels,W):

    #For the weights of the cut
    cluster_cut_weight = np.empty(len(set(labels)))
    ic_cut_weight = np.empty(len(labels))

    #For the weights within the cluster
    cluster_within_weight = np.empty(len(set(labels)))
    ic_within_weight = np.empty(len(labels))

    for k in set(labels):
        positions_out = [index for index, value in enumerate(labels) if value != k]
        positions_in = [index for index, value in enumerate(labels) if value == k]

        W_tri_out = np.triu(W,k=1)

        pairs_out = combinations(positions_out, 2)  # itertools.combinations ensures i < j
        pairs_in = combinations(positions_in, 2)
        pairs_in_out = product(positions_in,positions_out)

        # Remove the weights that never enter the cluster
        for i, j in pairs_out:
            W_tri_out[i, j] = 0
            
            
        W_tri_cut = np.triu(W_out,k=1) #matrix with zeros to everything not linked to cluster k
        W_tri_within = np.triu(W_out,k=1) #matrix with zeros to everything not linked to cluster k

        for i, j in pairs_in:
            W_tri_cut[i,j] = 0 #matrix with zeros to everything not linked to cluster k or completely inside cluster k. The only elements surviving are the ones that cut the cluster k 
        cluster_cut_weight[k]=np.sum(W_tri_cut)

        for i, j in pairs_cut:
            W_tri_within[i,j] = 0  #matrix with zeros to everything not linked to cluster k or with one vortex in and the other out. The only elements surviving are the ones completely inside the cluster k
        cluster_within_weight[k]=np.sum(W_tri_within)


    for i in range(len(labels)):
        ic_cut_weight[i] = cluster_cut_weight[labels[i]]
        ic_within_weight[i] = cluster_within_weight[labels[i]]

    return ic_cut_weight, ic_within_weight


```
